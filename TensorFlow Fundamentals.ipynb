1. What are the different data structures used in Tensorflow?. Give some examples

Ans:In TensorFlow, the primary data structure used is the Tensor. Tensors are multi-dimensional arrays that represent the data flowing through the computation graph. There are different types of tensors based on their dimensionality:

Scalar Tensor: A tensor with zero dimensions, representing a single value.
Example: tf.constant(5)

Vector Tensor: A tensor with one dimension, representing a list of values.
Example: tf.constant([1, 2, 3])

Matrix Tensor: A tensor with two dimensions, representing a 2D array of values.
Example: tf.constant([[1, 2], [3, 4]])

Higher-Dimensional Tensor: Tensors can have more than two dimensions for complex data.
Example: tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

Apart from tensors, TensorFlow also uses other data structures like constants, variables, and placeholders:

Constant: A constant tensor with fixed values that cannot be changed during the execution of the computation graph.
Example: tf.constant(3.14)

Variable: A variable tensor whose values can be modified during the execution of the computation graph. Variables are commonly used to represent model parameters.
Example: tf.Variable([1, 2, 3])

Placeholder: A placeholder tensor acts as a placeholder for input data that will be provided during the execution of the computation graph, typically during training or inference.
Example: tf.placeholder(tf.float32, shape=(None, 3))

These data structures form the foundation for defining and manipulating data in TensorFlow, enabling the creation and training of complex deep learning models.

 2. How does the TensorFlow constant differ from a TensorFlow variable? Explain with an example
 
 Ans: n TensorFlow, both constants and variables are used to store and handle data, but they have some key differences:

TensorFlow Constant:

A constant is a type of tensor whose value remains fixed during the execution of the computation graph.
Once a constant is created, its value cannot be changed.
Constants are typically used for storing fixed values like hyperparameters, mathematical constants, or other values that do not change during training.

TensorFlow Variable:

A variable is another type of tensor that holds values that can be modified during the execution of the computation graph.
Variables are often used to represent model parameters (e.g., weights and biases) that need to be updated during the training process to learn from data.

3. Describe the process of matrix addition, multiplication, and elementDwise operations in TensorFlow.

Ans: In TensorFlow, matrix operations are performed using tensors, which are multi-dimensional arrays similar to matrices. Let's describe the process of matrix addition, multiplication, and element-wise operations in TensorFlow:

Matrix Addition:
Matrix addition in TensorFlow is performed using the tf.add() function or the + operator. Both methods add corresponding elements of two tensors (matrices) together to create a new tensor with the same shape as the input tensors.
Example:

python
Copy code
import tensorflow as tf

##### Define two tensors (matrices)
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[5, 6], [7, 8]])

##### Perform matrix addition using tf.add()
C = tf.add(A, B)

##### The result C will be a new tensor with the same shape as A and B
##### Output: [[6, 8], [10, 12]]
Matrix Multiplication:
Matrix multiplication in TensorFlow is performed using the tf.matmul() function or the @ operator. It computes the dot product of two tensors according to the rules of matrix multiplication.
Example:

python
Copy code
import tensorflow as tf

#### Define two tensors (matrices)
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[5, 6], [7, 8]])

##### Perform matrix multiplication using tf.matmul()
C = tf.matmul(A, B)

##### The result C will be a new tensor with shape (2, 2)
##### Output: [[19, 22], [43, 50]]
Element-wise Operations:
Element-wise operations in TensorFlow perform operations element by element between two tensors of the same shape. Common element-wise operations include addition, subtraction, multiplication, division, exponentiation, and more.
Example:

python
Copy code
import tensorflow as tf

##### Define two tensors (matrices)
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[5, 6], [7, 8]])

###### Perform element-wise addition using the + operator
C = A + B

##### The result C will be a new tensor with the same shape as A and B
###### Output: [[6, 8], [10, 12]]

##### Perform element-wise multiplication using tf.multiply()
D = tf.multiply(A, B)

#### The result D will be a new tensor with the same shape as A and B
##### Output: [[5, 12], [21, 32]]
